```python
import matplotlib.pyplot as plt
```

```python
x=[1,2,3,4,5]
```

```python
y=[5,1,3,2,4]
y1=[1,5,2,4,3]
```

```python
plt.scatter(x,y)
# only markers
```

```python
plt.bar(x,y)
# bar chart as vertically
```

```python
plt.barh(x,y)
# bar chart as horizontally
```

```python
plt.bar(x,y,width=0.2,color=["red","Blue","green"])
# change width if its vertically
```

```python
plt.barh(x,y,height=0.2,color=["red","Blue","green"])
# use height for horizontally
```

```python
plt.pie(y)
# pie chart
```

```python
plt.pie(y,labels=["A","B","C","D","E"])
# add labels
```

```python
# plt.pie(y,labels=["A","B","C","D","E"],autopct="%i%%")
plt.pie(y,labels=["A","B","C","D","E"],autopct="%.2f%%",frame=True,shadow=True,explode=[0,0.2,0,0.4,0])
plt.show()

# for percentage ratio 
# have to use minium 2 %
# in 3 % give datatype int=(i),float=(f)
# to reduce length of float use .2 or .3
# frame = True to add frame
# shadow=True to add shadows
```

```python
import seaborn as sns
```

```python
data=sns.load_dataset("tips")
data
# seaborn has their own libraries 
# one of its libraries is "tips"
```

```python
plt.scatter(data["total_bill"],data["tip"]);
```

```python
sns.relplot(data=data,x="total_bill",y="tip",hue="sex");
# relplot=relation plot
# hue to saprate data here added sex that saprate data to male and female
```

```python
sns.relplot(data=data,x="total_bill",y="tip",hue="sex",style="smoker",size="time");
# sns.relplot(data=data,x="total_bill",y="tip",hue="sex",style="smoker",size="time",kind="line");
# stype and size used to saprate data in category
# kind="line" used to make chart of lines
```

```python
sns.jointplot(data=data,x="total_bill",y="tip")
# by default will joint satter and another chart
```

```python
sns.jointplot(data=data,x="total_bill",y="tip",kind="reg");
# ['scatter', 'hist', 'hex', 'kde' can add hue , 'reg', 'resid']
```

```python
sns.countplot(x=data["sex"],hue=data["smoker"])
# to count data
```

```python
sns.countplot(x=data["sex"],hue=data["smoker"])
# to count data
```
